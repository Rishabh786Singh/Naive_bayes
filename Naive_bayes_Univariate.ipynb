```python
#here we can see that time and result_of_treatment is highly correlated
feature=data['Time'].copy()
target=data['Result_of_Treatment'].copy()
```

```python
pos_prob=len(data[data['Result_of_Treatment']==1])/90#p(k=1)
neg_prob=len(data[data['Result_of_Treatment']==0])/90#p(k=0)
print('Probability of succesfull treatment :',pos_prob, " Probability of unsuccesfull treatment :",neg_prob)
```

```python
#we assume that our features follows normal distribution
def prior(x,mean,std):
    term1=-((x-mean)/std)**2
    return 0.399*(2.73**term1)/std
def posterior(x,pos_mean,pos_std,neg_mean,neg_std):
    global pos_prob, neg_prob
    num0=prior(x,pos_mean,pos_std)*pos_prob
    num1=prior(x,neg_mean,neg_std)*neg_prob
    return num0/(num0+num1),num1/(num0+num1)

def accuracy(y_predicted,y_actual):
    return 100*(len(target[y_predicted==y_actual])/len(y_actual))
```

```python
predicted=[]
pos_mean=np.mean(data[data['Result_of_Treatment']==1]['Time'])
neg_mean=np.mean(data[data['Result_of_Treatment']==0]['Time'])
pos_std=np.std(data[data['Result_of_Treatment']==1]['Time'])
neg_std=np.std(data[data['Result_of_Treatment']==0]['Time'])
for i in range(len(data)):
    positive, negative=posterior(feature.iloc[i],pos_mean,pos_std,neg_mean,neg_std)
    if(positive>=negative):
        predicted.append(1)
    else:
        predicted.append(0)
    
```

```python
accuracy(target,predicted)
```

```python
#just to check our accuracy
count=0
for i in range(len(data)):
    if(target.iloc[i]==predicted[i]):
        count=count+1
        
print(count/90)        
```
